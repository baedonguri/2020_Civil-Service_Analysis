```python
from konlpy.tag import Hannanum
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
```

```python
text = ''
with open('C:/PYTHON/시간대.txt', 'r') as f:
    text = f.read()

hannanum = Hannanum()
```

```python
nouns= hannanum.nouns(text)#명사
# morphs = hannanum.morphs(text) #문장

print(nouns)
```

```python
pos = hannanum.pos(text) #형태소 분석
print(pos)
```

```python
#명사 중에 길이가 1 이상인것만 추출
words = []
for n in nouns:
    if len(n)>1:
        words.append(n)
print(words)
```

```python
count = Counter(words) #빈도수 계산
most = count.most_common(100)

tags = {}
for n, c in most:
    tags[n] = c

tags
```

```python
wc = WordCloud(font_path = 'C:/PYTHON/NanumGothic.ttf',background_color='white', width=1200, height=800, scale=20, max_font_size=250)

gen = wc.generate_from_frequencies(tags)

plt.figure()
plt.imshow(gen, interpolation='bilinear')
wc.to_file('C:/PYTHON/주차관리과.png')
plt.close()
```
